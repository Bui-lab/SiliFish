```python
# pylint: disable=reportMissingModuleSource
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as patches
from CrossCorr.Const import SWIM_MODE_COLOR

colors = [SWIM_MODE_COLOR[1], SWIM_MODE_COLOR[3], SWIM_MODE_COLOR[5]]

data = pd.DataFrame({
    "SwimMode": pd.Categorical(["Slow", "Intermediate", "Fast"], 
                               categories=["Slow", "Intermediate", "Fast"]),
    #values are merged from 5 runs of slow/intermediate/fast scale and sequential simulations
    "Xmin": [19.03, 34.71, 56.99], 
    "Xmax": [36.00, 58.38, 78.42]
})
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)

fig, ax = plt.subplots(figsize=(8, 4))

for i, row in data.iterrows():
    rect = patches.Rectangle((row.Xmin, i - 0.4), row.Xmax - row.Xmin, 0.8, 
                             linewidth=1, edgecolor='black', facecolor=colors[i])
    ax.add_patch(rect)

ax.set_xlim(15, 85)
ax.set_ylim(-0.5, 2.5)
ax.set_yticks([0, 1, 2])
ax.set_yticklabels(data["SwimMode"])
ax.set_xlabel("TBF (Hz)")
ax.set_ylabel("Swim Mode")


ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
